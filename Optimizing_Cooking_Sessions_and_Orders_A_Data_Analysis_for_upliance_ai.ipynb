## 3. ***Data Wrangling***

### Data Wrangling Code

```python
# Write your code to make your dataset analysis ready.
# Handling Null Values in OrderDetails_data
# It is observed that when the order status is cancelled there is no rating given so to impute null values in ratings  with 0 is reasonable

OrderDetails_data.fillna(0, inplace= True)
OrderDetails_data.isnull().sum()
```

```python
# Merge UserDetails_data with OrderDetails_data on 'User ID'
merged_df = pd.merge(OrderDetails_data, UserDetails_data, on='User ID', how='inner')

# Merge the result with CookingSessions_data on 'Session ID'
final_df = pd.merge(merged_df, CookingSessions_data, on='Session ID', how='inner')

final_df.head()
```

```python
# Dropping duplicate columns
final_df.drop(columns=['User ID_y', 'Dish Name_y', 'Meal Type_y'], inplace= True)  # Keep the '_x' columns and drop '_y'
# Rename columns to make them more consistent
final_df.rename(columns={'User ID_x': 'User ID', 'Meal Type_x': 'Meal Type', 'Dish Name_x': 'Dish Name'}, inplace = True)
```

```python
final_df.head()
```

## ***4. Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables***

### User Demographics


#### Chart - Total Orders by Age Group

```python
# Chart - visualization code
# Creating age groups
final_df['Age Group'] = pd.cut(final_df['Age'], bins=[0, 20, 30, 40, 50, 100], labels=['0-20', '21-30', '31-40', '41-50', '51+'])

# Bar plot for Total Orders by Age Group
plt.figure(figsize=(8, 6))
sns.countplot(data=final_df, x='Age Group', palette='Set2')
plt.title('Total Orders by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Total Orders')
plt.show()
```

##### 2. What is/are the insight(s) found from the chart?

Most orders are from the 21-30 age group during dinner hours, emphasizing the need for targeted campaigns."


#### Chart - Users Over Time (Registration)

```python
# Chart - visualization code
plt.figure(figsize=(10, 6))
final_df['Registration Date'].value_counts().sort_index().plot(kind='line', marker='o', color='purple')
plt.title('Users Over Time (Registration)')
plt.xlabel('Registration Date')
plt.ylabel('Number of Users')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

##### 2. What is/are the insight(s) found from the chart?

The new users over time decreases which indicated the need to enlarging the customer base and promoting effective marketing

#### Chart - Top 5 Users by Total Spending

```python
# Grouping by User ID to calculate metrics
top_users = final_df.groupby('User Name').agg(
    Total_Orders=('Order ID', 'count'),
    Total_Spent=('Amount (USD)', 'sum')
).reset_index()

# Sorting users by Total Spent
top_users = top_users.sort_values(by='Total_Spent', ascending=False).head(5)  # Top 10 users

# Plotting the bar chart for Total Spent
plt.figure(figsize=(8, 6))
sns.barplot(x='User Name', y='Total_Spent', data=top_users, palette='Set2')
plt.title('Top 5 Users by Total Spending')
plt.xlabel('User Name')
plt.ylabel('Total Amount Spent (USD)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

##### 2. What is/are the insight(s) found from the chart?

The new users over time decreases which indicated the need to enlarging the customer base and promoting effective marketing
